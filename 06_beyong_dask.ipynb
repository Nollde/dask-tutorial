## Framework comparison
(inspired by [this blogpost](https://domino.ai/blog/spark-dask-ray-choosing-the-right-framework)):

### [Dask](https://www.dask.org)
- Developed 2015 by (the former) Anaconda
- Originally aimed to parallelize Data Science (NumPy, Pandas), now being developed as general-purpose library of parallel computing
- Pro:
    - Very pythonic
    - Gentle learning curve
- Con:
    - More focussed on data-science (but flexible)

### [Ray](https://www.ray.io)
- Developed 2017 @ UC Berkeley
- General low-level framework for parallelizing python code, focussed on ML applications
- Pro:
    - Very pythonic
    - Strong focus on ML applications
- Con:
    - No straight forward replacement of numpy & pandas
    - Moderate learning curve

### [Apache Spark](https://spark.apache.org)
- Developed 2009 @ UC Berkeley
- Unified Engine for large-scale data analytics
- Pro:
    - Industry-player
    - Large, mature community
- Con:
    - Industry-player
    - Not pythonic
    - Steep learning curve

### [Message Passing Interface (MPI)](https://www.open-mpi.org)
- Low-level standard for parallel and/or distributed computing
- Pro:
    - Huge flexibility and low-level controls
    - Huge speedups possible
    - Language Agnostic
- Con:
    - Very (!) steep learning curve

### General recommendation:
Use the right tool for the job and **have an infrastructure which is flexible enough for a mix-and-match approach**.
